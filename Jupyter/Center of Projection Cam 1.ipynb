# Finding the Center of Projection

```python
def Distance(p1: np.array, p2: np.array):
    x1, y1 = p1
    x2, y2 = p2
    return((x1-x2)**2 + (y1-y2)**2)**0.5

def InCenter(p1, p2, p3):
    # returns the position of a triangle's incenter
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    l1 = Distance(p2, p3)
    l2 = Distance(p1, p3)
    l3 = Distance(p1, p2)

    inCenter = (l1*p1 + l2*p2 + l3*p3) / (l1 + l2 + l3)
    
    
    # A = np.array([[1/(y3 - y2), -1/(x3-x2)],[1/(y3-y1), -1/(x3-x1)]])
    # print(f"A = {A}")
    # B = np.array([[x2/(y3-y2) - y2/(x3-x2)],[x1/(y3-y1)-y1/(x3-x1)]])
    # print(f"B = {B}")
    # x, y = np.linalg.inv(A) @ B
    return inCenter
```

```python
p1 = np.array([int(x1), int(y1)])
p2 = np.array([int(x2), int(y2)])
p3 = np.array([int(x3), int(y3)])
Cop = InCenter(p1, p2, p3)
print(f"Cop = {Cop}")
xCop, yCop = Cop
```
